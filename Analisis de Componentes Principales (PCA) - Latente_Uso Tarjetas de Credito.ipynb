```python
# Importar Bibliotecas

import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV como un DataFrame de pandas con ';' como separador

df = pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\tarjetacredito.csv', sep=';', index_col='ID')
```

```python
# Crear y ajustar el objeto PCA a los datos
# se ha terminado que 6 componentes explican el 99% de la variabilidad de los conjuntos de datos de tarjetas de credito
pca6 = PCA(n_components = 6)
```

```python
pca6.fit(df)
```

```python
# Matriz de correlación Variables Originales y Componentes (Latentes)
# Crearemos un DataFrame llamado df_pca6componentes, que contiene las cargas de los seis primeros componentes
# principales obtenidos del PCA. Estas cargas representan la correlación de cada variable original con los
# componentes y son clave para interpretar la contribución de cada variable a los componentes identificados.
```

```python
df_pca6componentes = pd.DataFrame(data=pca6.components_,
                                 columns = df.columns.values,
                                 index = ["Componente 1", "Componente 2", "Componente 3", "Componente 4", "Componente 5", "Componente 6"])
```

```python
df_pca6componentes
```

```python
pca6.components_
```

```python
# Mapa de Calor: Ayuda a identificar rápidamente qué variables contribuyen más a cada componente principal, facilitando la 
# interpretación de las dimensiones reducidas del PCA.

plt.figure(figsize = (20,10), dpi = 300)
mapa = sns.heatmap(df_pca6componentes,
                annot = True,
                cmap = "RdBu",
                vmin = -1,
                vmax =  1)
```

```python
#Para interpretar los resultados de un análisis de componentes principales (PCA) y etiquetar cada componente con una variable latente representativa, vamos a observar cómo las variables originales se correlacionan con los diferentes componentes. A partir de estas correlaciones, podemos asignar un "label" intuitivo a cada componente que resuma la naturaleza subyacente de la variable latente. Luego, se ofrece una descripción para cada componente.

Componente 1: "Capacidad Crediticia y Pago"
Descripción: Este componente está altamente correlacionado con el límite de crédito (0.689), el monto pagado (0.487), y en menor medida con el saldo (balance) (0.291). Es posible que este componente esté capturando la capacidad general del cliente para manejar crédito, así como su comportamiento de pago. Los clientes con altos límites de crédito y montos pagados probablemente reflejan una sólida salud financiera o al menos una gran capacidad de crédito.
Componente 2: "Patrón de Gasto en Compras"
Descripción: Este componente tiene una fuerte correlación con el total de compras (0.501) y la compra más grande realizada en un solo pago (0.381), lo que sugiere que está capturando el comportamiento del cliente en relación con las compras. Los clientes en este grupo son aquellos que probablemente hacen compras significativas y tienden a usar su tarjeta para realizar grandes transacciones.
Componente 3: "Uso de Adelanto de Efectivo y Deuda"
Descripción: Este componente está fuertemente correlacionado con el adelanto de efectivo (0.397) y el pago mínimo (0.639), lo que indica que está capturando un comportamiento relacionado con el uso de crédito en efectivo y una tendencia a hacer pagos mínimos. Este perfil podría describir a clientes que recurren a avances en efectivo con mayor frecuencia y que tienden a cumplir con los pagos mínimos, posiblemente como señal de estrés financiero.
Componente 4: "Comportamiento de Pago Total y Deuda"
Descripción: Está relacionado principalmente con el pago total (0.635) y con una correlación negativa en el uso de adelanto de efectivo (-0.565). Esto podría reflejar un grupo de clientes que pagan regularmente el total de su deuda, evitando el uso de efectivo, lo que indica un comportamiento financiero prudente y controlado.
Componente 5: "Inestabilidad Financiera y Bajos Saldos"
Descripción: Presenta una correlación negativa significativa con el saldo (balance) (-0.720) y las compras (-0.289). Podría interpretarse como un indicador de inestabilidad financiera o de clientes que no acumulan grandes saldos y no usan mucho su tarjeta de crédito, lo que refleja quizás una baja actividad financiera o un enfoque más conservador con el uso del crédito.
Componente 6: "Dependencia de Adelantos en Efectivo"
Descripción: Fuertemente correlacionado negativamente con el uso de adelantos de efectivo (-0.633) y relacionado con un bajo pago mínimo (-0.142), este componente parece capturar a clientes que dependen mucho de los adelantos de efectivo, posiblemente aquellos en situaciones más urgentes o de mayor riesgo financiero.
Resumen de Componentes Latentes:
Capacidad Crediticia y Pago: Clientes con altos límites de crédito, que realizan pagos considerables.
Patrón de Gasto en Compras: Clientes que hacen compras significativas y grandes transacciones en un solo pago.
Uso de Adelanto de Efectivo y Deuda: Clientes que recurren a adelantos en efectivo y tienden a hacer solo los pagos mínimos.
Comportamiento de Pago Total y Deuda: Clientes que pagan el total de su deuda y evitan el uso de adelantos en efectivo.
Inestabilidad Financiera y Bajos Saldos: Clientes con bajos saldos y poca actividad de compra.
Dependencia de Adelantos en Efectivo: Clientes que dependen mucho de adelantos de efectivo.
Este análisis proporciona una visión clara sobre diferentes perfiles de usuarios de tarjetas de crédito y su comportamiento financiero.
```
