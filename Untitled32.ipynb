```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV como un DataFrame de pandas con ';' como separador

df = pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\tarjetacredito.csv', sep=';', index_col='ID')
```

```python
df.describe()
```

```python
# Crear y ajustar el objeto PCA a los datos

pca = PCA()
```

```python
pca.fit(df)
```

```python
#Varianza explicada por cada componente
#Este array nos dice cuánta información (en términos de varianza de los datos) se captura con cada componente principal.
#Por ejemplo, si el primer valor de explained_variance_ratio_ es 0.475, esto significa que el primer componente principal
#explica el 47.5% de la varianza total en el conjunto de datos. Si el segundo valor es 0.178, el segundo componente explica 
#un 17.8% adicional de la varianza, y así sucesivamente.
```

```python
var_explicada = pca.explained_variance_ratio_
```

```python
var_explicada
```

```python
# Graficar la Varianza Explicada Acumulada

sns.set(style="darkgrid")
plt.figure(figsize=(15,7), dpi = 300)
plt.plot(range(1, len(var_explicada) + 1), var_explicada.cumsum(), marker = "o", linestyle = '-')
plt.title("Varianza Explicada por los Componentes")
plt.xlabel("Número de Componentes")
plt.ylabel("Varianza Explicada Acumulada")
plt.show()
```

```python
# Podemos interpretar varios puntos clave a partir de este gráfico:

# Crecimiento rápido inicial: La curva muestra un aumento rápido en la varianza explicada acumulada con los primeros componentes, lo que indica que estas primeras dimensiones capturan una cantidad significativa de información sobre los datos.

# Punto de inflexión: Alrededor del cuarto componente, la tasa de aumento en la varianza explicada comienza a estabilizarse, lo que sugiere que los componentes adicionales están agregando menos información.

# Meseta de la curva: Después de aproximadamente el sexto componente, la línea se vuelve casi horizontal, lo que implica que agregar más componentes no contribuye sustancialmente a la cantidad de varianza explicada en los datos.

# Selección de componentes para reducción de dimensiones: Basándose en este gráfico, uno podría elegir usar alrededor de cinco o seis componentes para la reducción de dimensiones, ya que estos parecen capturar la mayor parte de la información útil de los datos originales.

# Aplicabilidad para segmentación: La selección de un número reducido de componentes principales antes de aplicar técnicas de clustering, como K-Means, puede simplificar el análisis sin sacrificar demasiada información crítica, lo que facilita una segmentación de mercado más eficiente.

# En resumen, el gráfico es una herramienta de decisión para determinar el número óptimo de componentes principales a utilizar en análisis posteriores, como la segmentación de clientes o la visualización de datos multidimensionales.''' 
```
